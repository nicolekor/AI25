```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
import konlpy
from wordcloud import WordCloud 
import PIL
```

```python
moon = open('./data/speech_moon.txt', encoding='UTF-8').read()
moon
```

```python
# 정규화 하기

moon = re.sub('[^가-힣]', ' ', moon)
moon
```

```python
hannanum = konlpy.tag.Hannanum()
hannanum.nouns("대한민국의 영토는 한반도와 그 부속도서로 한다")
```

```python
#연설문에서 명사 추출하기
nouns = hannanum.nouns(moon)
nouns
```

```python
#데이터 프레임으로 변환

df_word = pd.DataFrame({'word' : nouns})
df_word
```

```python
#단어 빈도표 만들기 _ 글자 수 추가 p283

df_word['count'] = df_word['word'].str.len()
df_word
```

```python
# 두글자 이상 쓰인 단어만 남기기

df_word = df_word.query('count >= 2')
df_word.sort_values('count')
```

```python
#단어 빈도 구하기 p284

df_word = df_word.groupby('word', as_index=False).agg(n = ('word', 'count')).sort_values('n', ascending=False)
df_word
```

```python
# 단어 빈도 막대 그래프 만들기

top20 = df_word.head(20)
top20
```

```python
plt.rcParams.update({
    'font.family': 'Malgun Gothic',
    'figure.dpi': 120,
    'figure.figsize': [6.5, 6]
})


sns.barplot(data=top20, y='word', x='n')
```

```python
# p286 워드클라우드 만들기
# 한글폰트 설정

# font = 'DoHyeon-Regular.ttf'
font = 'C:/Users/pc04-18/Downloads/Do_Hyeon/DoHyeon-Regular.ttf'
```

```python
#데이터 프레임을 딕셔너리(key-value)로 변환

dic_word = df_word.set_index('word').to_dict()['n']
dic_word
```

```python
# p288 워드클라우드 만들기

wc = WordCloud(
    random_state=1234, 
    font_path=font,             # 여기서 font 변수도 정의돼 있어야 합니다
    width=400,
    height=400,
    background_color='white'
)
```

```python
#워드 클라우드 만들기
img_wordcloud = wc.generate_from_frequencies(dic_word)

# 워드 클라우드 출력하기
plt.Figure(figsize=(10, 10))
plt.axis('off')
plt.imshow(img_wordcloud)
```

```python
# WC 만들기

wc = WordCloud(random_state= 1234,
               font_path=font,
               width=400,
               height=400,
               background_color='white',
               mask= img)
```

```python
# p293 가장많이 사용된 단어 알아보기 _ 기사댓글불러오기

df = pd.read_csv('./data/news_comment_BTS.csv', encoding='UTF-8')
df.info()
```

```python
# 불필요한 문자 제거

df['reply'] = df['reply'].str.replace('[^가-힣]', ' ', regex=True)
df['reply'].head()
```

```python
# 명사추출하기 p294

kkma = konlpy.tag.Kkma()
```

```python
nouns = df['reply'].apply(kkma.nouns)
nouns
```

```python
# 단어 빈도표 만들기 p295

nouns = nouns.explode()
nouns
```

```python
# 두글자 이상인 단어만 뽑아내기

#데이터 프레임만들기
df_word = pd.DataFrame({'word' : nouns})

#글자 수 추가
df_word['count'] = df_word['word'].str.len()

#  두 글자 이상 단어만 남기기
df_word = df_word.query('count >= 2')
df_word
```

```python
# 빈도표 만들기 p296

df_word = df_word.groupby('word', as_index=False).agg(n = ('word', 'count')).sort_values('n', ascending=False)
df_word
```

```python
# 상위 20개 사용 빈도 단어들 추출

top20 = df_word.head(20)
top20
```

```python
# 가로세로 크기 설정

plt.rcParams.update({'figure.figsize': [6.5, 6]})

# 바차트로 만들어 보기
sns.barplot(data=top20, y='word', x='n')
```
